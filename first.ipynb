```python
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
import os
from dotenv import load_dotenv

load_dotenv()

options = webdriver.EdgeOptions()
options.add_experimental_option('prefs', {'download.default_directory' : 'C:\\FRANCIS_ECHESI',"download.directory_upgrade":True})
# options.add_argument("enable-managed-downloads")
# options.add_argument("--enable-managed-downloads")
# options.add_argument("enable-managed-downloads=true")
# options.add_argument("--enable-managed-downloads=true")
options.enable_downloads = True

driver = webdriver.Edge(options=options)

driver.get("https://www.jotform.com/")
driver.implicitly_wait(0.5)

login_button = driver.find_element(by=By.LINK_TEXT, value="Login")
login_button.click()
driver.implicitly_wait(2.0)

email_box = driver.find_element(by=By.NAME, value="email")
password_box = driver.find_element(by=By.NAME, value="pass")
signin_button = driver.find_element(by=By.ID, value="signinButton")

email_box.send_keys(os.getenv("email"))
password_box.send_keys(os.getenv("password"))
signin_button.click()
driver.implicitly_wait(5.0)
url = driver.command_executor._url
session_id = driver.session_id
cookies = driver.get_cookies()

with open('cookies.json', 'w') as file:
    json.dump(cookies, file)

with open('authinstance.json', 'w') as file:
    json.dump({'session_id': session_id, "url": url},file)
```
